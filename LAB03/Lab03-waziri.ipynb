```python
from skimage import io, filters, img_as_ubyte
from skimage.filters import median
from skimage.morphology import disk
from scipy.ndimage import uniform_filter
import matplotlib.pyplot as plt
```

```python
image = io.imread('image.jpg', as_gray=True)
plt.imshow(image, cmap='gray')
plt.axis('off')
plt.title('Original X-ray Image')
plt.show()
```

```python
image = img_as_ubyte(image)
```

```python
median_filtered = median(image, disk(2))
```

```python
mean_filtered = uniform_filter(median_filtered, size=(5, 5)) # 5x5 kernel
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 2)
plt.imshow(median_filtered, cmap='gray')
plt.title('median filtered img')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(1, 3, 3)
plt.imshow(mean_filtered, cmap='gray')
plt.title('median & mean filtered img')
plt.axis('off')
plt.show()
```

```python
io.imsave('filtered_image.jpg', mean_filtered)
```
