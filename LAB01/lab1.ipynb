```python
from matplotlib import pyplot as plt

zero = iio.imread(uri="eight.tif")
zero[1,2]= 1.0
zero[3,0]= 1.0

fig, ax = plt.subplots()
plt.imshow(zero)
print(zero)
```

```python

path = input("Enter the path to the image: ")
color = float(input("Enter the level of transparency(0-255): "))
image = iio.imread(uri=path)   
plt.imshow(image)
print(image)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        image[i,j, 2] = color 
fig, ax = plt.subplots()
plt.imshow(image)
plt.show()

patht = input("Enter the path to save the image: ")
iio.imwrite(patht,image)
print("Image saved successfully")
```

```python
import matplotlib.pyplot as plt
from skimage import exposure, io


image = io.imread('dog2.jpg')

# Apply gamma correction
gamma_corrected = exposure.adjust_gamma(image, gamma=0.4)

```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
ax = axes.ravel()

ax[0].imshow(image)
ax[0].set_title('original dog')

ax[1].imshow(gamma_corrected)
ax[1].set_title('gama corrected dog')

for a in ax:
    a.axis('off')

plt.tight_layout()
plt.show()
```
