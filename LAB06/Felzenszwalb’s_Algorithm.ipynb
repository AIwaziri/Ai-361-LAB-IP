```python
import skimage.segmentation
from matplotlib import pyplot as plt
from skimage.util import img_as_float
from skimage.segmentation import mark_boundaries
```

```python
img = plt.imread("date_cluster.jpg")
```

```python
res1 = skimage.segmentation.felzenszwalb(img, scale=500)
res2 = skimage.segmentation.felzenszwalb(img, scale=100)
```

```python
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(res1); ax1.set_xlabel("cluster=500")
ax2.imshow(res2); ax2.set_xlabel("cluster=100")
fig.suptitle("graph  segmentation using felzenszwalb method")
plt.tight_layout()
```

```python
img = img_as_float(img)
fig = plt.figure(figsize=(12, 5))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax1.imshow(mark_boundaries(img, res1)); ax1.set_xlabel("With cluster=500")
ax2.imshow(mark_boundaries(img, res2)); ax2.set_xlabel("With cluster=100")
fig.suptitle("the  boundaries")
```

### Lab Report

	This code takes an image called date_cluster.jpg and breaks it into smaller pieces using a method called felzenszwalb. It creates two versions of the segmented image: one with bigger chunks (using a scale of 500) and one with smaller chunks (using a scale of 100). Then, it shows both images side by side so you can compare how the image is divided with different settings. After that, the code highlights the borders of these chunks on the original image, making it easy to see where the divisions are. In short, its a way to split an image into parts and visualize the results.
