```python
import matplotlib.pyplot as plt
from plotly.express import imshow
from skimage import io, color, filters, measure, morphology
from skimage.draw import rectangle_perimeter
```

```python
image = io.imread('date_cluster.jpg')  
gray_image = color.rgb2gray(image)
```

```python
thresh_value = filters.threshold_otsu(gray_image) # Otsu's thresholding
binary_image = gray_image > thresh_value
```

```python
cleaned_image = morphology.remove_small_objects(binary_image, min_size=500)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(gray_image, cmap='gray')
ax[0].set_title("gray img")
ax[1].imshow(cleaned_image, cmap='gray')
ax[1].set_title("segmented img")
plt.show()
```
