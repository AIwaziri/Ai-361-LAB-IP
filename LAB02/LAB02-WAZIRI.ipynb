```python
import skimage
from skimage import data, io
import matplotlib.pyplot as plt
```

```python
image = io.imread('image.jpg')
```

```python
io.imshow(image, cmap='gray')
```

```python
#1.	Display a cropped color image of the model space shuttle in the image.
plt.imshow(image[0:200, 0:100])
```

```python
#2.	Display each color channel of image [0:200, 0:100] in grayscale.
new_img = image[0:200, 0:100]
img1 = new_img[:,:,0]
img2 = new_img[:,:,1]
img3 = new_img[:,:,2]

plt.subplot(131)
plt.imshow(img1, cmap='gray')
plt.subplot(132)
plt.imshow(img2, cmap='gray')
plt.subplot(133)
plt.imshow(img3, cmap='gray')
plt.show()
```
